```
import random
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt

n = 1000
radius = 5
x = 0
y = 0
z = 0
ran = random.uniform(-radius,radius)

def f(a):
    a=random.uniform(-1,1)
    a = a*ran
    return a

f_x =[]
f_y =[]
f_z =[]
i = 0

while True:
    x= f(x)
    y= f(y)
    z= f(z)
    if x**2+y**2+z**2 <= radius:
        f_x.append(x)
        f_y.append(y)
        f_z.append(z)
        i = i+1
    if i == n:
        break

fig = plt.figure()
ax = fig.gca(projection='3d')

particle = [10,0,0]
ax.plot(f_x, f_y, f_z)
E_f =[]

for i in range(n):
    d = sqrt((f_x[i]-10)**2+f_y[i]**2+f_z[i]**2)
    Ef =  1/(d**2)
    E_f.append(Ef)
```
